```python
import pandas as pd
import numpy as np
```

### Example 1

```python
df1 = pd.DataFrame(np.array([[2000,'A',100], 
                             [2001,'A',104],
                             [2002,'A',105],
                             [2000,'B',520],
                             [2001,'B',535],
                             [2002,'B',537],
                             [2000,'C',1020],
                             [2001,'C',1030],
                             [2002,'C',1040]]),
                   columns=['year', 'name', 'income'])

df1
```

```python
df1.pivot(index = 'name', columns = 'year', values = 'income')
```

### Example 2

```python
df2 = pd.DataFrame(np.array([[2000,'Ana','A',100,10], 
                             [2001,'Ana','A',104,11],
                             [2002,'Ana','A',105,12],
                             [2000,'Bob','B',520,23],
                             [2001,'Bob','B',535,24],
                             [2002,'Bob','B',537,25],
                             [2000,'Cici','C',1020,31],
                             [2001,'Cici','C',1030,32],
                             [2002,'Cici','C',1040,33]]),
                   columns=['year', 'fullname', 'name', 'income', 'age'])

df2
```

```python
df2A = df2.pivot(index = 'fullname', 
                       columns = 'year', 
                       values = ['income','age'])
df2A
```
