## Mapping education data to ONET SOC codes: An illustration

This python notebook searches into http://www.onetcodeconnector.org and assigns the relevant SOC codes. For example, a search for "plastic operator" can be done by putting in:
https://www.onetcodeconnector.org/find/result?s=plastic+operator

In this table, there are total of eight variables:
1. Score
2. Occupation
3. SOC code
4. Whether the search appears in the ONET SOC title  
5. Whether the search appears in the Lay titles
6. Whether the search appears in the ONET SOC description
7. Whether the search appears in the ONET SOC tasks
8. Whether the search appears in the detail work activities

```python
search_input = [w for w in re.split('\n',open('search_input.csv').read()) if not w=='']
```

```python
df = pd.DataFrame()

for search_phrase in search_input:
    df_temp = search_ONET_connector(search_phrase)
    df = df.append(df_temp, ignore_index=True)
    
df.to_csv('search_output.csv',index=False)
```
